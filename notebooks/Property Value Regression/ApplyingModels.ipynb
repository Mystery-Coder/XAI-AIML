### Trying out different models and checking the accuracy of the models.

#### Model: Random Forest Regressor

```python
# Using Random Forest Regressor

from sklearn.model_selection import train_test_split

import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor  
import joblib


num_df = pd.read_csv("../../data/Property Value Regression/train_cleaned.csv")  
X = num_df.drop("TARGET(PRICE_IN_LACS)",axis=1)
y = num_df["TARGET(PRICE_IN_LACS)"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.007, random_state=42)

model = RandomForestRegressor(n_estimators=100, random_state=42)

model.fit(X_train, y_train)


# plt.figure(figsize=(10, 6))
# plt.scatter(x_axis,y_test, label='Actual')
# plt.scatter(x_axis,model.predict(X_test), label='Predicted')
# plt.xlabel('Samples')
# plt.ylabel('Values')
# plt.title('Model Prediction vs Actual Values')
# plt.legend()
# plt.show()

model.score(X_test, y_test)

joblib.dump(model, 'random_forest_model.joblib')
print("Saved joblib")
```
